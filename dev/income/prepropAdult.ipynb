```python
# Delete features 3,7,10,11 and 13 on training data
train_df = pd.read_csv("/home/erick/Desktop/mcc20201/nnets/datasets/income/adult.data", header = None).drop([3, 7, 10, 11, 13],axis=1)
print(train_df.shape)
train_df[:3]
```

```python
# Delete features 3,7,10,11 and 13 on test data
test_df = pd.read_csv("/home/erick/Desktop/mcc20201/nnets/datasets/income/adult.test", header = None).drop([3,7,10, 11,13],axis=1)
print(test_df.shape)
test_df[:3]
```

### One-hot-encoding of categorical variables

```python
training_data = pd.get_dummies(pd.concat([train_df,test_df])).iloc[:32561]
test_data = pd.get_dummies(pd.concat([train_df,test_df])).iloc[32561:]
print(training_data.shape)
training_data[:3]
```

### Min-Max scaling of continuos variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
training_data = pd.DataFrame(scaler.fit_transform(training_data), columns=training_data.columns)
test_data = pd.DataFrame(scaler.fit_transform(test_data), columns=test_data.columns)
training_data[:3]
```

```python
np.savetxt("train_income.dat", training_data, delimiter=" ",  fmt='%f9')
np.savetxt("test_income.dat", test_data, delimiter=" ",  fmt='%f9')
```
