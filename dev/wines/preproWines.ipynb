```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/erick/Desktop/mcc20201/nnets/wine.data", header = None)
df.head()
```

```python
y = df.iloc[0:,0]
df = df.drop([0], axis=1)
```

```python
y.head()
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
df.head()
```

```python
a = np.ndarray(shape=(178,3)) 
for i in range(59):
    a[i] = [1,0,0]
for i in range(70):
    a[i+59] = [0,1,0]
for i in range(49):
    a[i+59+70] = [0,0,1]
print(len(a))
a[:5]
```

```python
data = np.concatenate((df,a), axis=1)
data[0]
```

```python
np.savetxt("data.csv", data, delimiter=" ",  fmt='%f9')
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.20, random_state=0)
```

```python
np.savetxt("train_data.csv", train_data, delimiter=" ",  fmt='%f9')
np.savetxt("test_data.csv", test_data, delimiter=" ",  fmt='%f9')
```
